```python
import pandas as pd
```

```python
beta_df = pd.read_csv("beta_df.csv")
```

```python
# Chaos Round assumption
market_crash_pct = -10

# Start from beta table
chaos_stock = beta_df.copy()

# Expected stock loss using beta
chaos_stock["Expected Loss if NIFTY50 drops 10% (%)"] = (
    chaos_stock["Beta vs NIFTY50"] * market_crash_pct
)

# Round for display
chaos_stock["Expected Loss if NIFTY50 drops 10% (%)"] = (
    chaos_stock["Expected Loss if NIFTY50 drops 10% (%)"].round(2)
)

# Sort from worst expected loss to least loss
chaos_stock = chaos_stock.sort_values(
    "Expected Loss if NIFTY50 drops 10% (%)"
).reset_index(drop=True)

print(chaos_stock)
```

```python
print("\nMost exposed stocks:")
print(chaos_stock.head())

print("\nSafest-refuge candidates:")
print(chaos_stock.tail())
```

```python
portfolio_a = pd.read_csv("portfolio_a.csv")
```

```python
# Portfolio A beta = weighted average beta
portfolio_a_beta = (
    portfolio_a.merge(beta_df, on="Ticker", how="left")
    .assign(weighted_beta=lambda df: df["Weight"] * df["Beta vs NIFTY50"])
)["weighted_beta"].sum()

portfolio_a_expected_loss = portfolio_a_beta * market_crash_pct

print("Portfolio A Beta:", round(portfolio_a_beta, 3))
print("Portfolio A Expected Loss if NIFTY50 drops 10% (%):", round(portfolio_a_expected_loss, 2))
```

```python
portfolio_b = pd.read_csv("portfolio_b.csv")
```

```python
# Portfolio B beta = weighted average beta
portfolio_b_beta = (
    portfolio_b.merge(beta_df, on="Ticker", how="left")
    .assign(weighted_beta=lambda df: df["Weight"] * df["Beta vs NIFTY50"])
)["weighted_beta"].sum()

portfolio_b_expected_loss = portfolio_b_beta * market_crash_pct

print("Portfolio B Beta:", round(portfolio_b_beta, 3))
print("Portfolio B Expected Loss if NIFTY50 drops 10% (%):", round(portfolio_b_expected_loss, 2))
```

```python
chaos_portfolio = pd.DataFrame({
    "Portfolio": ["Portfolio A", "Portfolio B"],
    "Portfolio Beta": [round(portfolio_a_beta, 3), round(portfolio_b_beta, 3)],
    "Expected Loss if NIFTY50 drops 10% (%)": [
        round(portfolio_a_expected_loss, 2),
        round(portfolio_b_expected_loss, 2)
    ]
})

print(chaos_portfolio)
```
